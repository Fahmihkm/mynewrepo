```python
# Distribution graphs (histogram/bar graph) of column data
def plotgraph(dt, ngs, ngpr): # ngs = ngraphshown , ngpr = ngraphperrow
    nunique = dt.nunique()
    dt = dt[[col for col in dt if nunique[col] > 1 and nunique[col] < 50]] # For displaying purposes, pick columns that have between 1 and 50 unique values
    nRow, nCol = dt.shape
    columnNames = list(dt)
    ngr = (nCol + ngpr - 1) / ngpr # ngr = ngraphrow
    plt.figure(num = None, figsize = (6 * ngpr, 8 * ngr), dpi = 80 )
    for i in range(min(nCol, ngs)):
        plt.subplot(ngr, ngpr, i + 1)
        columnDt = dt.iloc[:, i]
        if (not np.issubdtype(type(columnDt.iloc[0]), np.number)):
            valueCounts = columnDt.value_counts()
            valueCounts.plot.bar()
        else:
            columnDt.hist()
        plt.ylabel('counts')
        plt.title(f'{columnNames[i]} (column {i})')
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
        
```

```python
plotgraph(data, 10, 1)

```
